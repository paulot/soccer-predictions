# Sequence History Experiment: Higher-Order MCMC Transitions

In our baseline MCMC model, we assume a **first-order Markov chain**—meaning the probability of the next state (pass success or destination) depends *only* on the current state (current zone, player on the ball, etc.).

However, football is a sequential game. A team's passing success is often influenced by the **buildup sequence**:
*   Did the possession start deep, drawing the opponent forward?
*   Was the previous pass a progressive ball that broke the defensive lines, leaving the opponent disorganized?
*   How many passes have occurred in this possession (defensive fatigue vs. attacking momentum)?

This experiment evaluates whether leveraging the **past N states** as features improves our ability to predict the **pass success rate (Outcome Model)**, and identifies the optimal window size $N$.

---

## 1. Experimental Design

For each pass in a possession, we will extract:
1. **Current Features:** `start_zone_x`, `start_zone_y`, `passer_accuracy`, `opp_defensive_rate`.
2. **Historical Features (for $i \in \{1, .. , N\}$):**
   *   `prev_i_zone_x`, `prev_i_zone_y`: Where did the previous $i$-th pass start?
   *   `prev_i_success`: Was the previous $i$-th pass completed?
   *   *If there are fewer than $i$ previous passes in the current possession, these features are padded with a default value of `-1` (representing "Start of Possession").*

We will evaluate **$N = 0$ (Baseline)**, **$N = 1$**, **$N = 2$**, and **$N = 3$** using a Random Forest Classifier and compare:
*   **Validation Accuracy**
*   **Log Loss** (probabilistic calibration)
*   **Feature Importance** of the historical features.

```python
import os
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, log_loss
import statsbombpy.sb as sb

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Locate raw event caches
raw_dir = "../data/raw_events"
print(f"Raw event directory: {raw_dir} (Exists: {os.path.exists(raw_dir)})")
```

## 2. Feature Extraction Pipeline with Sequence History

We will write a function to load raw match events, sort them chronologically, group by possession, and extract the sequence of passes.

```python
def parse_location(loc_str):
    if pd.isna(loc_str): return None
    try:
        return ast.literal_eval(loc_str)
    except:
        return None

def get_zone_coords(loc):
    if not loc or len(loc) < 2:
        return None, None
    x, y = loc[0], loc[1]
    zone_x = min(int(x / 20), 5)
    zone_y = min(int(y / 16), 4)
    return zone_x, zone_y

def extract_sequence_features(limit_matches=50):
    """Extracts passing sequences from local raw event files."""
    match_files = [f for f in os.listdir(raw_dir) if f.endswith('.csv')][:limit_matches]
    
    all_passes = []
    
    # Load profiles for contextual features
    df_profiles = pd.read_csv("../data/statsbomb_player_profiles.csv").set_index('player')
    player_accuracy = df_profiles['accuracy'].to_dict()
    player_prog = df_profiles['progressive_ratio'].to_dict()
    
    df_def = pd.read_csv("../data/team_defensive_profiles.csv")
    defensive_profiles = {}
    for _, row in df_def.iterrows():
        t, z, r = row['team'], row['zone'], row['defensive_rate']
        if t not in defensive_profiles: defensive_profiles[t] = {}
        defensive_profiles[t][z] = r
        
    df_mgr = pd.read_csv("../data/manager_profiles.csv").set_index('manager')
    mgr_directness = df_mgr['directness'].to_dict()
    mgr_width = df_mgr['width'].to_dict()
    
    print(f"Processing {len(match_files)} matches...")
    for filename in match_files:
        match_id = filename.split('.')[0]
        df_events = pd.read_csv(os.path.join(raw_dir, filename))
        
        # Sort chronologically
        df_events = df_events.sort_values(by=['period', 'timestamp'])
        
        # Filter for passes
        df_passes = df_events[df_events['type'] == 'Pass'].copy()
        if df_passes.empty: continue
        
        # Parse locations
        df_passes['loc_parsed'] = df_passes['location'].apply(parse_location)
        
        # Group by possession
        for poss_id, group in df_passes.groupby('possession'):
            poss_passes = group.sort_values(by=['period', 'timestamp']).to_dict(orient='records')
            
            for i, p in enumerate(poss_passes):
                loc = p['loc_parsed']
                curr_x, curr_y = get_zone_coords(loc)
                if curr_x is None: continue
                
                # Determine outcome
                outcome = 1 if not pd.isna(p.get('pass_outcome')) else 0  # 1 = Turnover, 0 = Success
                
                # Contextual features
                passer = p.get('player')
                accuracy = player_accuracy.get(passer, 0.75)
                prog_ratio = player_prog.get(passer, 0.20)
                
                opp_team = p.get('team') # Simple fallback
                zone_name = f"zone_{curr_x}_{curr_y}"
                def_rate = defensive_profiles.get(opp_team, {}).get(zone_name, 5.0)
                
                # Build history features (N=3)
                history = {}
                for n in range(1, 4):
                    if i - n >= 0:
                        prev_p = poss_passes[i - n]
                        prev_loc = prev_p['loc_parsed']
                        px, py = get_zone_coords(prev_loc)
                        p_outcome = 1 if not pd.isna(prev_p.get('pass_outcome')) else 0
                        
                        history[f'prev_{n}_zone_x'] = px if px is not None else -1
                        history[f'prev_{n}_zone_y'] = py if py is not None else -1
                        history[f'prev_{n}_success'] = 1 - p_outcome
                    else:
                        # Padding for start of possession
                        history[f'prev_{n}_zone_x'] = -1
                        history[f'prev_{n}_zone_y'] = -1
                        history[f'prev_{n}_success'] = -1
                
                record = {
                    'start_zone_x': curr_x,
                    'start_zone_y': curr_y,
                    'passer_accuracy': accuracy,
                    'passer_progressive_ratio': prog_ratio,
                    'opp_defensive_rate': def_rate,
                    'outcome': outcome,
                    **history
                }
                all_passes.append(record)
                
    return pd.DataFrame(all_passes)

df_seq = extract_sequence_features(limit_matches=80)
print(f"Compiled dataset with {df_seq.shape[0]:,} passes.")
df_seq.head()
```

## 3. Model Training & Comparison

We will evaluate four models, increasing the history window size $N$ from 0 to 3.

```python
# Split into Train/Val
X_all = df_seq.drop(columns=['outcome'])
y_all = df_seq['outcome']

X_train_raw, X_val_raw, y_train, y_val = train_test_split(X_all, y_all, test_size=0.2, random_state=42)

# Define feature sets for different N
base_features = ['start_zone_x', 'start_zone_y', 'passer_accuracy', 'passer_progressive_ratio', 'opp_defensive_rate']

feature_sets = {
    "N=0 (Baseline)": base_features,
    "N=1 (Previous Pass)": base_features + ['prev_1_zone_x', 'prev_1_zone_y', 'prev_1_success'],
    "N=2 (Previous 2 Passes)": base_features + [
        'prev_1_zone_x', 'prev_1_zone_y', 'prev_1_success',
        'prev_2_zone_x', 'prev_2_zone_y', 'prev_2_success'
    ],
    "N=3 (Previous 3 Passes)": base_features + [
        'prev_1_zone_x', 'prev_1_zone_y', 'prev_1_success',
        'prev_2_zone_x', 'prev_2_zone_y', 'prev_2_success',
        'prev_3_zone_x', 'prev_3_zone_y', 'prev_3_success'
    ]
}

results = []
models = {}

for name, feats in feature_sets.items():
    print(f"Training model: {name} (Features: {len(feats)})...")
    
    X_tr = X_train_raw[feats]
    X_va = X_val_raw[feats]
    
    rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
    rf.fit(X_tr, y_train)
    
    # Predictions
    preds = rf.predict(X_va)
    probs = rf.predict_proba(X_va)
    
    acc = accuracy_score(y_val, preds)
    loss = log_loss(y_val, probs)
    
    results.append({
        'Model': name,
        'Accuracy': acc,
        'Log Loss': loss
    })
    models[name] = (rf, feats)
    
df_res = pd.DataFrame(results)
df_res
```

## 4. Visualizing Performance Gains

Let's plot the Accuracy and Log Loss as a function of the history window size $N$.

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

color = 'tab:blue'
ax1.set_xlabel('Sequence History Window Size (N)')
ax1.set_ylabel('Validation Accuracy', color=color)
sns.lineplot(x='Model', y='Accuracy', data=df_res, marker='o', ax=ax1, color=color, sort=False)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Log Loss (Lower is Better)', color=color)
sns.lineplot(x='Model', y='Log Loss', data=df_res, marker='s', ax=ax2, color=color, sort=False)
ax2.tick_params(axis='y', labelcolor=color)

plt.title("Impact of Sequence History on Pass Success Prediction")
fig.tight_layout()
plt.show()
```

## 5. Feature Importance Analysis (For N=3)

Let's look at the feature importance of the $N=3$ model to see how much weight the classifier places on the buildup history versus the current state.

```python
rf_3, feats_3 = models["N=3 (Previous 3 Passes)"]
importances = rf_3.feature_importances_

df_imp = pd.DataFrame({
    'Feature': feats_3,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

sns.barplot(x='Importance', y='Feature', data=df_imp, palette='viridis')
plt.title("Feature Importance for N=3 Sequence History Model")
plt.xlabel("Gini Importance")
plt.show()

df_imp
```

## 6. Conclusions & MCMC Implications

### Key Questions Addressed:
1. **Does sequence history improve predictions?**
   * If validation accuracy increases and Log Loss decreases as $N$ increases, it proves that pass success is not strictly first-order Markovian. Buildup momentum and defensive displacement play a measurable role.
2. **What is the optimal window size $N$?**
   * Typically, we expect a large jump from $N=0$ to $N=1$, and diminishing returns for $N=2$ and $N=3$. This helps us balance model complexity versus predictive power.

### Implications for MCMC Simulation:
* If $N \ge 1$ offers significant gains, we should update our transition engine to maintain a small **possession history buffer** of length $N$ during the simulation, rather than evaluating transitions purely on the current state.
